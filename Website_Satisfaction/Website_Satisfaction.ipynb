I am taking data from a Website Satisfaction Survey to analyze what matters most to people of different demographics browsing the internet.

Data: https://www.kaggle.com/datasets/hayriyigit/website-satisfaction-survey

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

print('done')
```

```python
df = pd.read_csv('survey.csv', index_col=0)
df
```

```python
df['Avg_typography'] = np.nan
df['Avg_color'] = np.nan
df['Avg_content'] = np.nan
df['Avg_interactivity'] = np.nan
df['Avg_navigation'] = np.nan
df['Avg_satisfaction'] = np.nan
df['Avg_trust'] = np.nan
df['Avg_loyalty'] = np.nan
df
```

```python
df['Avg_typography'] = (df['q1'] + df['q2'] + df['q3']) / 3
df['Avg_color'] = (df['q4'] + df['q5']) / 2
df['Avg_content'] = (df['q6'] + df['q7'] + df['q8']) / 3
df['Avg_interactivity'] = (df['q9'] + df['q10'] + df['q11'] + df['q12'] + df['q13']) / 5
df['Avg_navigation'] = (df['q14'] + df['q15'] + df['q16'] + df['q17'] + df['q18']) / 5
df['Avg_satisfaction'] = (df['q19'] + df['q20'] + df['q21']+ df['q22']) / 4
df['Avg_trust'] = (df['q23'] + df['q24']) / 2
df['Avg_loyalty'] = (df['q25'] + df['q26']) / 2
df.head()
```

```python
df['age_quartile'] = pd.qcut(df['age'], 4, ['1Q','2Q','3Q','4Q'])
df['typography_q'] = pd.qcut(df['Avg_typography'], 4, ['1Q','2Q','3Q','4Q'])
df['color_q'] = pd.qcut(df['Avg_color'], 4, ['1Q','2Q','3Q','4Q'])
df['content_q'] = pd.qcut(df['Avg_content'], 4, ['1Q','2Q','3Q','4Q'])
df['interactivity_q'] = pd.qcut(df['Avg_interactivity'], 4, ['1Q','2Q','3Q','4Q'])
df['navigation_q'] = pd.qcut(df['Avg_navigation'], 4, ['1Q','2Q','3Q','4Q'])
df['satisfaction_q'] = pd.qcut(df['Avg_satisfaction'], 4, ['1Q','2Q','3Q','4Q'])
df['trust_q'] = pd.qcut(df['Avg_trust'], 4, ['1Q','2Q','3Q','4Q'])
df['loyalty_q'] = pd.qcut(df['Avg_loyalty'], 4, ['1Q','2Q','3Q','4Q'])
```

Which features of a website are most likely to improve the Satisfaction, Trust, and Loyalty of a customer?

```python
sns.heatmap(df.corr(),linewidths=.05, annot=True)
```

The 2 most important factors to improving customer satisfaction are how easy it is to navigate the website and how interactive it is to customer needs.

The 2 most important factors for improving trust are navigation and existing loyalty.

The 2 most important factors for improving loyalty are trust and how easy it is to navigate and personalize.

Does content of the website have an impact on customer satisfaction for differing age groups?

```python
sns.lineplot(x='Avg_content',y='satisfaction_q', hue='age_quartile', data=df)
plt.gca().invert_yaxis()
```

The content of the website matters a lot more to peopler in the younger Age Quartiles(1 and 2) vs people in the older range quartile (3 and 4)

Does color satisfaction of the website have an impact on customer satisfaction for differing age groups?

```python
sns.lineplot(x='color_q',y='satisfaction_q', hue='age_quartile', data=df)
plt.gca().invert_yaxis()
```

Similar to above, the color satisfaction of a website matters a lot more to the younger generation vs the older generation.

I want to see, do loyalty, trust, and satisfaction have a linear relationship?

```python
sns.lineplot(x='loyalty_q',y='satisfaction_q', data=df)
plt.gca().invert_yaxis()
```

```python
sns.lineplot(x='trust_q',y='satisfaction_q', data=df)
plt.gca().invert_yaxis()
```

Chart above shows that more or less, loyalty, trust and satisfaction have a linear relationship. The more loyal someone is to a website or trusts a website, the more satisfied they are.

Do people in different age groups tend to be more satisfied than people in other age groups?

```python
sns.lineplot(x='age_quartile',y='Avg_satisfaction', data=df)
plt.gca().invert_yaxis()
```

People in older age groups tend to have a higher average satisfaction rate than people in younger age groups.

How does the Platform the customer is using have an impact on overall customer satisfaction?

```python
sns.lineplot(x='platform',y='Avg_satisfaction',hue='gender', data=df)
```

This chart is quite interesting to look at in the exploratory analysis phase. Males tend to have higher satisfaction using mobile devices vs desktop, while the exact opposite is true for females.

# Conclusion

There is a ton of insightful information that can be gained using this data set. 

I found that Trust, Loyalty, and Satisfaction are all linearly related, as one increases so do the other two, on average.

I also found that the age of the perso browing has an impact on how likely they are to be satisfied. With people in higher age quartile, tend to have a higher satisfaction rate vs people in lower age quartiles.

I found that color scheme of a website matters a lot more to the younger generation than the older generation.

Among all age groups, we found that ease of navigation and interactivity of the website have the biggest impact on satisfaction, even slightly above the content the website is providing.
